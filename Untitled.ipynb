```python
import csv
lol = (list(csv.reader(open('C:/Users/Dan/Desktop/bayer/irradiance.txt', 'rb'), delimiter='\t')))


wavelength=[]
intensity=[]
for aa in lol:
    wavelength.append(aa[0])
    intensity.append(aa[1])

intensity=np.asarray(intensity, dtype=float)
sns.set_context("poster")
normalizedIntensity= (intensity- np.min(intensity))/(np.max(intensity)-np.min(intensity))
plt.plot(wavelength, normalizedIntensity)
plt.ylabel('Transmittance')
plt.xlabel('Wavelength (nm)')
plt.savefig('C:/Users/Dan/Desktop/bayer/'+'hamatsuSpectrum', dpi=1000)
```

```python
redSpectrumHolder = []
wavelengthHolder =[]
for spectrumFiles in os.listdir('C:/Users/Dan/Desktop/bayer/new_spectrum'):
    f = io.open('C:/Users/Dan/Desktop/bayer/new_spectrum/'+spectrumFiles, "rb")

    data=f.read()[-6404096:]

    #print data[:4] 
    assert data[:4] == 'BRCM'
    data = data[32768:]

    data = np.fromstring(data, dtype=np.uint8)
    data=data.reshape((1952, 3264))[:1944, :3240]
    data = data.astype(np.uint16) << 2
    for byte in range(4):
        data[:, byte::5] |= ((data[:, 4::5] >> ((4 - byte) * 2)) & 0b11)
    data = np.delete(data, np.s_[4::5], 1)
    rgb = np.zeros(data.shape + (3,), dtype=data.dtype)
    rgb[1::2, 0::2, 0] = data[1::2, 0::2] # Red
    rgb[0::2, 0::2, 1] = data[0::2, 0::2] # Green
    rgb[1::2, 1::2, 1] = data[1::2, 1::2] # Green
    rgb[0::2, 1::2, 2] = data[0::2, 1::2] # Blue
    
    
    redMatrix =rgb[1000:1204:1,1000:1204:1, 0]
    averageHolder = redMatrix.sum()
    #[spectrumFiles[11:-4],
    redSpectrumHolder.append(averageHolder)
    wavelengthHolder.append(spectrumFiles[11:-4])
sns.set_context("poster")

matrixSize = redMatrix.size

fig = plt.figure()
redSpectrumHolder=np.array(redSpectrumHolder, dtype=float)
redSpectrumHolderNormalized = (redSpectrumHolder-redSpectrumHolder.min())/(redSpectrumHolder.max()-redSpectrumHolder.min())
plt.plot(wavelengthHolder, redSpectrumHolderNormalized, 'ro-', label='Red Filter')
plt.ylabel('Transmittance')
plt.xlabel('Wavelength (nm)')
plt.savefig('C:/Users/Dan/Desktop/bayer/'+'redSpectrum', dpi=1000)
```
